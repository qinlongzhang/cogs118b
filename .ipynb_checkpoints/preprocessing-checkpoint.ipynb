```python
import os
from skimage.io import imread
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

WORKING_DIR = './images/val/'

def read_image(image_path):
    files = os.listdir(image_path)
    images = []
    
    for fi in files:
        image = imread(os.path.join(image_path, fi))
        images.append(image)
        
    return np.array(images)
```

```python
all_images = read_image(WORKING_DIR)
plt.imshow(all_images[0])
```

```python
def cut_images(all_images):
    new_images = []
    masks = []
    
    for image in all_images:
        if image.shape == (256, 512, 3):
            newImage = image[:, 0:256, :]
            oldmask = image[:, 256:512, :]
            
            mask = handle_mask(oldmask)
            
            new_images.append(newImage)
            masks.append(mask)
            
    return np.array(new_images), np.array(masks)

def handle_mask(mask):
    
    if (mask.shape == (256, 256, 3)):
        threshold = 30
        
        lowerbound = np.array((128 - threshold, 63 - threshold, 127 - threshold), dtype=np.uint8, ndmin=1)
        upperbound = np.array((128 + threshold, 63 + threshold, 127 + threshold), dtype=np.uint8, ndmin=1)
        
        mask = cv.inRange(mask, lowerbound, upperbound)
        
    return mask

new_images, masks = cut_images(all_images)

for i in range(new_images.shape[0]):
    cv.imwrite(WORKING_DIR + 'img/' + str(i) + '.jpg', new_images[i])
    cv.imwrite(WORKING_DIR + 'mask/' + str(i) + '.jpg', masks[i])
```
